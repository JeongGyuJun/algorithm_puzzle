```
#QUIZ_15 계단에 서서 수다 떨기 

#문제 : 10단짜리 계단에서 같은 방식으로 이동했을 때 두 사람이 같은 단에 서는 방법은 몇 가지인지 구해 보세요.


import sys

A = 0
B = 10

su =0

def stairs(A, B):
  global su
  count = 0
  #print(su)
  if A == B:
    return 1
  if A > B:
    return 0

  for i in range(1, 5):
    for j in range(1, 5):
      #su +=1
      count += stairs(A+i, B-j)

  return count


if __name__=='__main__':
  sys.setrecursionlimit(1000)
  result = stairs(A, B)
  print('총 같은 단에 서있는 경우의 수 :', result)
```

```
# QUIZ_17 30인 31각에 도전!

#문제 : 30인을 한 줄로 세우는 경우 몇 가지 방법이 있는지 구해 보세요. 남녀의 배열 방법만 생각하는 것으로 하고, 누가 어느 위치에 서느지는 무시하기로 합니다.
# 예를 들어 4인(4인 5각)의 경우 [그림3]과 같이 8가지가 있습니다.

input_data = int(input("몇인 몇각에 도전하시겠어요? "))

count = 0

def in_gag(a, b):
  global count
  count += 1
  if count == 30:
    return a+b
  return in_gag(b, a+b) 
  
result = in_gag(1, 1)
print('{}인 {}각의 경우의 수: {}가지'.format(input_data, input_data+1, result))
```

```
#QUIZ_20 수난의 파사드 마방진 

#문제 : 이 마방진을 사용하여 다음 조건으로 덧셈한 결과 그 합이 같은 조합을 가장 많이 가지는 값(합)을 구해 보세요.
#조건
#서로 더할 수 있는 것은 가로, 세로, 대각선에 한정하지 않음
#서로 더하는 수의 개수는 4개에 한정하지 않음.

from itertools import combinations

mabangjin = [1, 14, 14, 4, 11, 7, 6, 9, 8, 10, 10, 5, 13, 2, 3, 15]
num_sum = 0
mabangjin_sum = []
count_index = []

for j in range(133):
  count_index.append(0)

for i in range(1, len(mabangjin)+1):
  mabangjin_sum.append([sum(list(j)) for j in combinations(mabangjin, i)])

for u in mabangjin_sum:
  for v in range(len(u)):
    k = u.pop()
    count_index[k] += 1

max_num = max(count_index)

print('덧셈한 결과 그 합이 같은 조합을 가장 많이 가지는 값(합) :', count_index.index(max_num))
```

```
#QUIZ_21 배타적 논리합으로 만드는 삼각형

#문제 : 위로부터 순서대로 배치해 나갈 때 2,014번째의 '0'이 출력되는 것은 몇 번째 단계인지 구해 보세요.

count = 0
line = 1
row = [1]

while count < 2014 :
  new_row = [1]
  for i in range(len(row) - 1):
    xor = row[i]^row[i+1]
    new_row.append(xor)

    if xor == 0 :
      count += 1

  new_row.append(1)
  #print(new_row)
  line += 1
  row = new_row

print('2014번째의 0이 출력되는 라인 :', line, '개')
```
