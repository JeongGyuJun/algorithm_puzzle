```
# QUIZ_09 남녀 불균형
# m+w C M


man = 20
woman = 10

def combination(move, one_way):
  first = move
  second = one_way

  for i in range(1, one_way):
    data = move - i
    first *= data
  for j in range(1, one_way):
    data2 = one_way - j
    second *= data2
  if second ==0 or first == 0:
    return 0
  result = first // second
  return result

print(combination(7, 4))
  

def unbalance(m, w):
    Total = 0
    move = m-1 + w
    print("move, m: ", move, m)
    Total += combination(move-1, m-1)
    print('Total :', Total)

    for i in range(1, w-1):
        for j in range(1, m-1):
            i_2 = woman - i
            j_2 = man - j
          
            if i == j or i_2 == j_2:
                print("i, j, i-1 :", i+1, j+1, i)
                data = combination(i+j+2, i)
                data = data
                print('data :', data)
                Total -= data
            else :
                pass
                
    return Total

print(unbalance(man, woman))
```

```
#QUIZ_10 룰렛의 최댓값

from collections import deque

european = deque([0, 32, 15, 19, 4, 21, 2, 25, 17, 34, 6, 27, 13, 36, \
            11, 30, 8, 23, 10, 5, 24, 16, 33, 1, 20, 14, 31, 9, 22, 18, 29, 7, 28, 12, 35, 3, 26])

american = deque([0, 28, 9, 26, 30, 11, 7, 20, 32, 17, 5, 22, 34, 15, 3, \
            24, 36, 13, 1, 0, 27, 10, 25, 29, 12, 8, 19, 31, 18, 6, 21, 33, 16, 4 ,23, 35, 14, 2])


n = int(input('연속하는 개수를 지정하기.'))
print('입력 묶음 수 :', n)

def roulette(n) :
  Compare = 0
  european_max = 0
  american_max = 0

  for i in range(len(european) + 1):
    european_sum = 0
    american_sum = 0
    for k, j in enumerate(range(n)):
      eu_data = european.popleft()
      am_data = american.popleft()
      
      european_sum += eu_data
      american_sum += am_data

      european.append(eu_data)
      american.append(am_data)
      
      i += 1
    
    if european_max < european_sum :
      #print(euoropean_max, european_sum)
      european_max = european_sum
    if american_max < american_sum :
      #print(american_max, american_sum)
      american_max = american_sum
    if european_sum < american_sum :
      #print(european_sum,'<' ,american_sum)
      Compare += 1

  print('유로피안 스타일에서의 합이 아메리칸 스타일에서의 합보다도 작아지는 수:', Compare)
  print('유로피안 최댓값:', european_max)
  print('아메리칸 최댓값:', american_max)

roulette(n)
```

```
#QUIZ 13 복면산을 만족하게 하는 것은 몇 가지일까?
#조건 : 각 문자열에서 최상위 문자는 0이 들어가면 안된다.

from itertools import permutations # permutations -> 순열, combinations -> 조합  - 알아두기!

count = 0
for R, E, A, D, W, I, T, L, K, S in permutations(range(0, 9 + 1)):
  if R ==0 or W ==0 or T == 0 or S == 0: continue

  READ = R * 1000 + E * 100 + A * 10 + D * 1
  WRITE = W * 10000 + R * 1000 + I * 100 + T * 10 + E * 1
  TALK = T *1000 + A * 100 + L * 10 + K * 1

  SKILL = S * 10000 + K * 1000 + I * 100 + L * 10 + L * 1

  RWT = READ + WRITE + TALK # 합계
  
  if RWT == SKILL:
    print("READ + WRITE + TALK = SKILL = RWT")
    print(READ, "+", WRITE, "+", TALK, "=", SKILL, "=", RWT)
    count += 1

print('복면산을 만족하게 하는 전체 개수 :', count, '개')
```
