```
#QUIZ_1 앞뒤가 같은 10진수 만들기

#문제 : 10진수, 2진수, 8진수 그 어느 것으로 표현하여도 대칭수가 되는 수 중, 10진수의 10이상에서의 최솟값을 구해 보세요.

num = 10

while True:
  num_10 = str(num)
  num_8 = oct(num).replace("0o", "")
  num_2 = bin(num).replace("0b", "")

  if num_10[::] == num_10[::-1] and num_8[::] == num_8[::-1] and num_2[::] == num_2[::-1] :
    print("num_10 = ", num_10)
    print("num_8 = ", num_8)
    print("num_2 = ", num_2)
    break

  num += 1

```

```
#QUIZ_2 수열의 사칙 연산

#완성된 식을 계산한 결과 '원래 수를 거꾸로 나열한 숫자'가 되는 것.

#1,000 ~ 9,999 중에서 앞의 조건을 만족하는 수를 구해 보세요.

import re

select_location = ["*", ""]

for i in range(1000, 9999 + 1):
  digit = str(i)
  for j in range(2):
    for u in range(2):
      for v in range(2):
        computation = digit[3] + select_location[j] + digit[2] + select_location[u] \
              + digit[1] + select_location[v] + digit[0]
        computation = re.sub(r"0(\d+)", r"\1", computation)
        
        if len(computation) > 4 :
          commit = eval(computation)
          if i == commit : print("결과 : ", i)
```

```
#QUIZ_03 카드를 뒤집어라!

#문제 : 뒤집을 카드가 더는 없을 때 뒷면이 위를 향한 카드의 번호를 모두 구해 보세요.

data = [False]

data = data * 101

for i in range(1, 101):
  sum = i
  while True:
    if sum >= 101 : break

    data[sum] = not data[sum]

    sum += i

for i in range(101):
  if data[i] == True :
    print(i, data[i], sep = ", ")
```

```
#QUIZ_04 막대자르기

#문제 : n = 20, m = 3일 때의 횟수를 구해 보세요.

def bar_clipping(length, personnel, current):
  print(length, personnel, current)
  if current >= personnel :
    return 0
  elif current < length :
    print(length, personnel, current * 2)
    return 1 + bar_clipping(length, personnel, current * 2)
  else :
    print(length, personnel, current + length)
    return 1 + bar_clipping(length, personnel, current + length)

print(bar_clipping(3, 20, 1))
print(bar_clipping(5, 100, 1))
```

```
#QUIZ_06 콜라스 추측

#문제 : 10,000 이하의 짝수 중 앞의 2나 4와 같이 '처음의 수로 돌아가는 수'가 몇 개 있는지 구해 보세요.

#원리 - 자연수 n에 대해
# 1. n이 짝수인 경우, n을 2로 나눈다.
# 2. n이 홀수인 경우, n에 3을 곱해 1을 더한다.

#원리 수정 - 초깃값이 짝수면 맨 처음에만 n에 3을 곱하여 1을 더하는 것에서 시작하기로 한다.

def colas_conjecture(natural_number):
    num = natural_number
    #print(num, end='->')
    num = num * 3 + 1
    #print(num, end='->')
    while natural_number != num and num != 1 :
        if num % 2 == 0 : 
            num = num // 2
            #print(num, end='->')
        else :
            num = num * 3 + 1
            #print(num, end='->')
    #print()
    return num

Total = 0

for i in range(2, 10000+1, 2):
  colas_value = colas_conjecture(i)
  
  if i == colas_value: Total += 1

print('총 개수 : ', Total)
```

```
#QUIZ_07 날짜의 2진수 변환

#문제 : 연원일을 YYYYMMDD의 8자리 정수로 나타내었을 때 2진수로 변환하여 거꾸로 나열한 다음 다시 
# 10진수로 되돌렸을 때 원래 날짜와 같은 날짜가 되는 것을 찾아보세요.
# 기간은 지난 도쿄 올림픽(1964년 10월 10일)부터 다음 도쿄 올림픽(2020년 7월 24일 예정)으로 하겠습니다.

from datetime import datetime, timedelta

start = datetime.strptime("1964-10-10", "%Y-%m-%d")
end = datetime.strptime("2020-07-24", "%Y-%m-%d")
step = timedelta(days=1)

while start<=end :
  start_bin = bin(int(start.strftime("%Y%m%d"))).replace("0b","")
  
  if start_bin[::] == start_bin[::-1] :
    print(start)

  start += step
```
